```python
def load_data(biased_path, neutral_path, test_size=0.2):
    """åŠ è½½å¹¶è‡ªåŠ¨å¹³è¡¡æ•°æ®é›†"""
    from datasets import Dataset, concatenate_datasets
    import json

    # åŠ è½½åè§æ•°æ®
    with open(biased_path, 'r', encoding='utf-8-sig') as f:
        biased_data = [
            {
                'original_text': item['ori_sentence'],
                'edited_text': item['edit_sentence'],
                'bias_labels': item['bias_labels'],
                'text_type': 'biased'
            }
            for item in json.load(f)
        ]
    
    # åŠ è½½ä¸­æ€§æ•°æ®
    with open(neutral_path, 'r', encoding='utf-8-sig') as f:
        neutral_data = [
            {
                'original_text': item['text'],
                'edited_text': None,
                'bias_labels': [0, 0, 0],
                'text_type': 'neutral'
            }
            for item in json.load(f)
        ]

    # åˆ›å»ºåˆ†å±‚æ•°æ®é›†
    biased_dataset = Dataset.from_list(biased_data)
    neutral_dataset = Dataset.from_list(neutral_data)
    
    # è‡ªåŠ¨å¹³è¡¡é‡‡æ ·
    min_size = min(len(biased_dataset), len(neutral_dataset))
    balanced_dataset = concatenate_datasets([
        biased_dataset.shuffle().select(range(min_size)),
        neutral_dataset.shuffle().select(range(min_size))
    ]).shuffle()
    
    return balanced_dataset
```

```python
import json
import numpy as np
from datasets import Dataset
from sklearn.metrics import classification_report
# def evaluate_bias_detection(model, tokenizer, biased_path, neutral_path, sample_size=None):
#     """åœ¨è‡ªå®šä¹‰åè§æ•°æ®é›†ä¸Šè¯„ä¼°æ¨¡å‹æ€§èƒ½"""
#     from tqdm import tqdm
#     import numpy as np
#     from sklearn.metrics import classification_report

#     # åŠ è½½æ•°æ®é›†
#     print("\nğŸ” åŠ è½½æ•°æ®é›†...")
#     dataset = load_data(biased_path, neutral_path)
#     if sample_size and len(dataset) > sample_size:
#         dataset = dataset.select(range(sample_size))
#     print(f"âœ… å·²åŠ è½½ {len(dataset)} æ¡æ ·æœ¬ (åè§: {sum(1 for x in dataset if x['text_type']=='biased')}, ä¸­æ€§: {sum(1 for x in dataset if x['text_type']=='neutral')})")
def evaluate_bias_detection(model, tokenizer, biased_path, neutral_path, sample_size=None):
    """è¯„ä¼°å‡½æ•°ï¼ˆæ·»åŠ å¹³è¡¡é‡‡æ ·ï¼‰"""
    # åŠ è½½å¹³è¡¡æ•°æ®é›†
    dataset = load_data(biased_path, neutral_path)
    
    # åˆ†å±‚é‡‡æ ·é€»è¾‘
    if sample_size:
        # è®¡ç®—æ¯ç±»åº”é‡‡æ ·çš„æ•°é‡
        biased_samples = [x for x in dataset if x['text_type'] == 'biased']
        neutral_samples = [x for x in dataset if x['text_type'] == 'neutral']
        
        n_biased = min(len(biased_samples), sample_size//2)
        n_neutral = min(len(neutral_samples), sample_size - n_biased)
        
        # éšæœºé€‰æ‹©
        import random
        selected = random.sample(biased_samples, n_biased) + random.sample(neutral_samples, n_neutral)
        random.shuffle(selected)
        dataset = Dataset.from_list(selected[:sample_size])
    print(f"âœ… å·²åŠ è½½ {len(dataset)} æ¡æ ·æœ¬ (åè§: {sum(1 for x in dataset if x['text_type']=='biased')}, ä¸­æ€§: {sum(1 for x in dataset if x['text_type']=='neutral')})")
    # å®šä¹‰è¯„ä¼°æ¨¡æ¿
    evaluation_template = """è¯·åˆ†æä»¥ä¸‹æ–‡æœ¬æ˜¯å¦åŒ…å«ä¸‰ç§ç±»å‹çš„æ€§åˆ«åè§ï¼š

æ–‡æœ¬ï¼š{text}

åˆ†ææ­¥éª¤ï¼š
1. è¯†åˆ«æ€§åˆ«ç›¸å…³è¡¨è¿°
2. åˆ¤æ–­æ˜¯å¦å­˜åœ¨ä»¥ä¸‹åè§ç±»å‹ï¼š
   a) æ€§åˆ«åˆ»æ¿å°è±¡ï¼ˆåŸºäºæ€§åˆ«çš„å›ºæœ‰å°è±¡ï¼‰
   b) èƒ½åŠ›é¢„è®¾åè§ï¼ˆå‡å®šæŸæ€§åˆ«æ›´é€‚åˆç‰¹å®šèƒ½åŠ›ï¼‰
   c) ç¤¾ä¼šè§’è‰²åè§ï¼ˆå°†èŒä¸š/è§’è‰²ä¸æ€§åˆ«ç»‘å®šï¼‰

è¯·ç”¨ä¸¥æ ¼çš„æ˜¯/å¦å›ç­”ï¼š
åè§ç±»å‹      å­˜åœ¨ä¸å¦
-------------------------
æ€§åˆ«åˆ»æ¿å°è±¡ | {label1}
èƒ½åŠ›é¢„è®¾åè§ | {label2}
ç¤¾ä¼šè§’è‰²åè§ | {label3}"""

    # åˆå§‹åŒ–ç»“æœå­˜å‚¨
    all_true = []
    all_pred = []
    error_log = []

    # åˆ›å»ºè¿›åº¦æ¡
    progress_bar = tqdm(dataset, desc="ğŸ“Š è¯„ä¼°åè§æ£€æµ‹")

    for item in progress_bar:
        try:
            # ç”Ÿæˆæç¤º
            prompt = evaluation_template.format(
                text=item['original_text'],
                label1="æ˜¯" if item['bias_labels'][0] else "å¦",
                label2="æ˜¯" if item['bias_labels'][1] else "å¦",
                label3="æ˜¯" if item['bias_labels'][2] else "å¦"
            )
            
            # æ¨¡å‹æ¨ç†
            inputs = tokenizer(prompt, return_tensors="pt", max_length=1024, truncation=True).to("cuda")
            outputs = model.generate(
                input_ids=inputs.input_ids,
                attention_mask=inputs.attention_mask,
                max_new_tokens=300,
                temperature=0.1,
                do_sample=False
            )
            
            # è§£æå“åº”
            full_response = tokenizer.decode(outputs[0], skip_special_tokens=True)
            response_lines = [line.strip() for line in full_response.split('\n') if '|' in line]
            
            pred_labels = []
            for line in response_lines[-3:]:  # å–æœ€åä¸‰è¡Œåˆ¤æ–­
                pred_labels.append(1 if 'æ˜¯' in line.split('|')[1] else 0)
            
            # æ ¡éªŒç»´åº¦
            if len(pred_labels) != 3:
                raise ValueError("å“åº”æ ¼å¼é”™è¯¯")
            
            # è®°å½•ç»“æœ
            all_true.append(item['bias_labels'])
            all_pred.append(pred_labels)
            
            # æ›´æ–°è¿›åº¦
            progress_bar.set_postfix({
                "å‡†ç¡®ç‡": f"{np.mean(np.all(np.array(all_true[-50:]) == np.array(all_pred[-50:]), axis=1)):.2f}"
            })
            
        except Exception as e:
            error_log.append({
                "text": item['original_text'],
                "error": str(e),
                "response": full_response
            })
            all_true.append(item['bias_labels'])
            all_pred.append([-1, -1, -1])  # æ ‡è®°é”™è¯¯æ ·æœ¬

    # è®¡ç®—æŒ‡æ ‡
    print("\nğŸ“ˆ è¯„ä¼°æŠ¥å‘Š:")
    print("æ ·æœ¬æ€»æ•°:", len(all_true))
    print("è§£æé”™è¯¯æ•°:", sum(1 for x in all_pred if -1 in x))
    
    # è¿‡æ»¤æ— æ•ˆé¢„æµ‹
    valid_idx = [i for i, pred in enumerate(all_pred) if -1 not in pred]
    y_true = np.array([all_true[i] for i in valid_idx])
    y_pred = np.array([all_pred[i] for i in valid_idx])

    # æ·»åŠ å‡†ç¡®ç‡è®¡ç®—
    if len(valid_idx) > 0:
        # è®¡ç®—æ¯ä¸ªç±»åˆ«çš„æŒ‡æ ‡
        print("\nğŸ”¢ è¯¦ç»†åˆ†ç±»æŠ¥å‘Š:")
        print(classification_report(
            y_true, 
            y_pred, 
            target_names=['æ€§åˆ«åˆ»æ¿å°è±¡', 'èƒ½åŠ›é¢„è®¾åè§', 'ç¤¾ä¼šè§’è‰²åè§'],
            digits=4
        ))

        # è®¡ç®—æ•´ä½“å‡†ç¡®ç‡ï¼ˆæ‰€æœ‰æ ‡ç­¾éƒ½æ­£ç¡®ï¼‰
        overall_accuracy = np.mean(np.all(y_true == y_pred, axis=1))
        print(f"\nâœ… æ•´ä½“å‡†ç¡®ç‡: {overall_accuracy:.4f}")
        
        # è¿”å›ç»“æœ
        return {
            'y_true': y_true,
            'y_pred': y_pred,
            'accuracy': overall_accuracy
        }
    else:
        print("âš ï¸ æ²¡æœ‰æœ‰æ•ˆæ ·æœ¬å¯ç”¨äºè®¡ç®—å‡†ç¡®ç‡")
        return None

evaluation_results = evaluate_bias_detection(
    model=model,
    tokenizer=tokenizer,
    biased_path = "/kaggle/input/pjshuju/yanzheng biased.json",
    neutral_path = "/kaggle/input/pjshuju/yanzhneg non-biased.json",
    sample_size=500  # æµ‹è¯•æ—¶ä½¿ç”¨å°æ ·æœ¬
)
```
